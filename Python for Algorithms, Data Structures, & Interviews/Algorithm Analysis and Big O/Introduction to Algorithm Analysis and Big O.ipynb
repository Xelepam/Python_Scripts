```python
def sum1(n):
    final_sum = 0
    
    for x in range(n+1):
        final_sum += x
        
    return final_sum
```

```python
sum1(10)
```

```python
def sum2(n):
    return (n * (n + 1))/2
```

```python
sum2(10)
```

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

#Setting up runtime comparisons 
n = np.linspace(1, 10, 100)
labels = ['Constant', 'Logarithmic', 'Linear', 'Log Linear', 'Quadratic', 'Cubic', 'Exponential']
big_o = [np.ones(n.shape), np.log(n), n, n*np.log(n), n**2, n**3, 2**n]

#Plot Setup
plt.figure(figsize = (12,10))
plt.ylim(0, 50)

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label = labels[i])
    

plt.legend(loc = 0)
plt.ylabel('Relartive Runtime')
plt.xlabel('n')
```

```python
#O(1) Constant

def func_constant(values):
    print(values[0])
```

```python
lst = [1, 2, 3]
func_constant(lst)
```

```python
#O(n) Linear

def func_lin(lst):
    
    for val in lst:
        print(val)
```

```python
func_lin(lst)
```

```python
#O(n^2) Quadratic

def func_quad(lst):
    
    for item_1 in lst:
        for item_2 in lst:
            print(item_1, item_2)
```

```python
lst = [1, 2, 3, 4, 5]
func_quad(lst)
```

```python
def print_once(lst):
    
    for val in lst:
        print(val)
```

```python
print_once(lst)
```

```python
def print_2(lst):
    
    for val in lst:
        print(val)
        
    for val in lst:
        print(val)
```

```python
print_2(lst)
```

```python
def comp(lst):
    
    print(lst[0]) #O(1)
    
    
    ##### O(n/2) . . . O( 1/2 * n)
    midpoint = int(len(lst)/2)
    
    for val in lst[:midpoint]:
        print(val)
    #####
    
    
    for x in range(10): #O(10)
        print('Hello World!')
```

```python
comp(lst) #O(1 + n/2 + 10) or O(n) 
```

```python
def matcher(lst, match):
    
    for item in lst:
        if item == match:
            return True
        
    return False
```

```python
matcher(lst, 1) #Best case is O(1)
```

```python
matcher(lst, 11) #Worst case is O(n) 
```

```python
def create_list(n):
    
    new_list = []
    
    for num in range(n):
        new_list.append('new')
        
    return new_list
```

```python
create_list(10)
```

```python
def printer(n):
    
    for x in range(10): #Time Complexity is O(n)
        print('Hello World!') #Space Complexity is O(1)
```
